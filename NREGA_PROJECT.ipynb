# AGGREGATIONS BY STATE AND DISTRICT

```python
import pandas as pd

# Get a list of unique state names in the DataFrame
unique_states = data_new['state_name'].unique()

# Initialize an empty dictionary to store the sums for each state
state_sums = {}

# Loop through each state
for state in unique_states:
    # Create a boolean mask to filter rows for the current state
    state_mask = data_new['state_name'] == state
    # Use the mask to select rows for the current state
    state_data = data_new[state_mask]
    # Calculate the sum for each column, excluding 'state_name'
    column_sums = state_data.drop(columns='state_name').sum()
    # Store the sums in the dictionary using the state name as the key
    state_sums[state] = column_sums

# Convert the dictionary of sums to a DataFrame
sums_df = pd.DataFrame(state_sums).T.reset_index()
sums_df = sums_df.rename(columns={'index': 'state_name'})

# Print the resulting DataFrame
print(sums_df)
```

# R-squared (R2):

R-squared is a measure of how well the predicted values match the variance in 
the actual values. It ranges from 0 to 1, where 1 indicates a perfect fit.

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

# EXPLANATION 
### R-squared (also known as the coefficient of determination), is 0.9939.
### R-squared is a statistical measure of the proportion of the variance in the dependent variable (target) 
### that is explained by the independent variables (features) in the model. 

### An R-squared value close to 1 (in this case, 0.9939) indicates that the model explains a very high proportion 
### of the variance in the target variable. 
### Approximately 99.39% of the variability in the dependent variable is accounted for by the independent variables
### included in the model.

## Model Fit:

### The closer the R-squared value is to 1, the better the model fits the data. In this case the model is capturing a 
### significant portion of the variability in the target variable, suggesting a strong fit.

## Predictive Power:

### A high R-squared suggests that the model is effective at predicting the target variable based on the features.
### The model's predictions align closely with the actual values.
